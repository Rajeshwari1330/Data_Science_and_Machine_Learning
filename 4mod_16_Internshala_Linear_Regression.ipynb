NOTE: Item_Outlet_Sales is our target variable, we need to predict this variable using all these variables

NOTE: we can see that target variable is continuous in nature, so this is a regression problem. That's why we are performing here Linear Regression.

```python
# making the train dataset 

train = data[0:7999]
```

NOTE: To build a model, we require 2 types of dataset, Training dataset and Test dataset.

Training dataset: we will train our model.

Test dataset: Here we make a predictions and checks the results.

NOTE: Training dataset and Test dataset is that the Test dataset doesn't have the target variable. But in this we will be having the taeget variable. 

```python
# making the test dataset

test = data[8000:]
```

```python
test.head()
```

```python
#Training dataset
# only Independent variables
# dropping the target variable i.e. Item_Outlet_Sales, so that we can left with the independent variables.
# here axis=1 specifies that drop from the column.

x_train = train.drop('Item_Outlet_Sales', axis=1)
```

```python
x_train.head()
```

```python
# Dependent variable

y_train = train['Item_Outlet_Sales']
```

```python
y_train.head()
```

```python
# Test Dataset

x_test = test.drop('Item_Outlet_Sales', axis=1)
```

```python
x_test.head()
```

```python
true_p = test['Item_Outlet_Sales']
```

```python
true_p.head()
```

```python
# importing library

from sklearn.linear_model import LinearRegression
```

```python
# creating an object of LinearRegression as lreg

lreg = LinearRegression()
```

```python
# fit() function is used to fit the model

lreg.fit(x_train,y_train)
```

NOTE: It accepts only int or float i.e. numeric value, it doesn't accept the strings values.

NOTE: For this we create a dummification

```python
# create a dummies for train dataset

x_train = pd.get_dummies(x_train)
```

```python
x_train.head()
```

```python
x_train.shape
```

```python
# create a dummies for test dataset too.

x_test = pd.get_dummies(x_test)
```

```python
x_test.head()
```

```python
x_test.shape
```

```python
# Now run our model and check.

lreg.fit(x_train,y_train)
```

NOTE: still we are getting an error, we need to impute the missing values, let's fix that

```python
# Filling the missing values as zero for train and test dataset.

x_train.fillna(0, inplace=True)
```

```python
x_test.fillna(0, inplace=True)
```

```python
# Now run our model and check

lreg.fit(x_train,y_train)
```

NOTE: We have successfully fitted our model now.

```python
# Once we fitted our model, the next step is to predict our model.
# Now predict on the dataset.

pred=lreg.predict(x_test)
```
