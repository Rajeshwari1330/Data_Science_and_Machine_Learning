```python
# kmeans is the object here, and cluster value is 2, means the dataset is divided into two parts

kmeans = KMeans(n_clusters = 2)
```

```python
# using fit function

kmeans.fit(data)
```

```python
# prediction

pred = kmeans.predict(data)
```

```python
pred
```

NOTE: Here 0 is one cluster, and 1 is another cluster.

```python
pd.Series(pred).value_counts()
```

* How do we decide whether our cluster is good or not? we use inertia. 
* inertia returns the sum of square distance within cluster.
* It tells that how close the values are to the cluster centroid.
* Larger the value of these sum within square, the more spread out the cluster is, i.e the cluster is not more compact.
* we should have as low as value of this inertia.

```python
kmeans.inertia_
```

```python
# score() function
# returns the negative value of inertia.
# This is similar to inertia but a reverse sign.

kmeans.score(data)
```

NOTE: Load of value of inertia better the clustering is, shown in as follows

```python
# Store the inertia in the list called SSE.

SSE = []
```

```python
# creating a 20 clusters

for cluster in range(1,20):
    kmeans = KMeans(n_jobs = -1, n_clusters = cluster)
    kmeans.fit(data)
    SSE.append(kmeans.inertia_)
```

```python
frame = pd.DataFrame({'Cluster':range(1,20), 'SSE':SSE})
```

```python
frame.head()
```

NOTE: As we gone increasing the cluster value, the SSE keep on decreasing.

```python
# Optimal cluster value, generally done by using elbow-curve
# Cluster is on x-axis, and SSE will be in y-axix

plt.figure(figsize=(12,6))
plt.plot(frame['Cluster'], frame['SSE'], marker='o')
```

NOTE:   * Generally Kmeans used for distance metrix, we would want the distance to be on equal scale.
        * We will transform a data using StandardScalar module under the preprocessing package.
        * Scale the values such that they have 0 Mean and standard deviation of 1, everything will have the same scale.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

data_scaled = scaler.fit_transform(data)
```

```python
pd.DataFrame(data_scaled).describe()
```

```python
# creating a list for scaled dataset

SSE_scaled = []
```

```python
# calculate the inertia for scaled dataset

for cluster in range(1,20):
    kmeans = KMeans(n_jobs = -1, n_clusters = cluster)
    kmeans.fit(data_scaled)
    SSE_scaled.append(kmeans.inertia_)
```

```python
# check the graph now

frame_scaled = pd.DataFrame({'Cluster':range(1,20), 'SSE':SSE_scaled})
plt.plot(frame_scaled['Cluster'], frame_scaled['SSE'], marker='o')
plt.xlabel("Clusters")
plt.ylabel("SSE")
```

NOTE: In this graph we can see here, there are 4-5 clusters, and in the above graph there are 6-7 clusters.
    * Calculate this clusters from the above till ideal value.

```python
kmeans = KMeans(n_jobs = -1, n_clusters = 4)
kmeans.fit(data_scaled)
pred = kmeans.predict(data_scaled)
```
